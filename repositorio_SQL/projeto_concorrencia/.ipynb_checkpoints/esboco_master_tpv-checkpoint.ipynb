```r
library(readr)
library(dplyr)
library(ggplot2)
```

```r
base = read_csv('base_diferenca_desembolso_IF.csv')
```

```r
base = base %>% mutate(flag_ap = case_when(IF_adj %in% c('Banco do Brasil','Cooperativas','Itaú','Bradesco','Caixa','Outros') ~ 0, TRUE ~ 1))
```

```r
resumo <- base %>%
  filter(!is.na(diferenca_limite),
         !is.na(month_year),
         !is.na(flag_ap)) %>%
  mutate(flag_ap = factor(flag_ap)) %>%
  group_by(month_year, flag_ap) %>%
  summarise(
    perc_acima_0 = mean(diferenca_limite > 0) * 100,
    mediana_diferenca = median(diferenca_limite),
    .groups = "drop"
  ) %>%
  mutate(
    month_year = factor(month_year, levels = sort(unique(month_year)))
  )
```

```r
ggplot(resumo, aes(
  x = month_year,
  y = perc_acima_0,
  color = flag_ap,
  group = flag_ap
)) +
  geom_line() +
  geom_point() +
  labs(
    x = "month_year",
    y = "% diferenca_limite > 0",
    color = "flag_ap",
    title = "Percentual de diferenca_limite acima de 0 por mês"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

```r
ggplot(resumo, aes(
  x = month_year,
  y = mediana_diferenca,
  color = flag_ap,
  group = flag_ap
)) +
  geom_line() +
  geom_point() +
  labs(
    x = "month_year",
    y = "Mediana de diferenca_limite",
    color = "flag_ap",
    title = "Mediana de diferenca_limite por mês"
  ) +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
